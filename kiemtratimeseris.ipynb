```python
import itertools
# Hàm để tìm các tham số tốt nhất cho mô hình ARIMA
def find_best_arima_params(train):
    p = range(0, 3)
    d = range(0, 3)
    q = range(0, 3)
    pdq = list(itertools.product(p, d, q))
    best_score = float("inf")
    best_params = None
    for param in pdq:
        try:
            model = ARIMA(train, order=param)
            results = model.fit()
            mse = mean_squared_error(train, results.fittedvalues)
            if mse < best_score:
                best_score = mse
                best_params = param
        except:
            continue
    return best_params
```

```python
# Lựa chọn các cột cần phân tích
columns_to_analyze = ['truong_1', 'truong_2', 'truong_3', 'truong_4', 'truong_5']

# Kết quả phân tích sẽ được lưu trong dictionary
results = {}

# Lặp qua từng cột để huấn luyện mô hình và dự đoán
for column in columns_to_analyze:
    print(f'Processing column: {column}')
    
    # Chọn cột dữ liệu để phân tích chuỗi thời gian
    time_series_data = data[['date', column]]
    
    # Loại bỏ các nhãn trùng lặp
    time_series_data = time_series_data.drop_duplicates(subset='date')
    
    # Đặt cột 'date' làm chỉ số và thiết lập tần số
    time_series_data.set_index('date', inplace=True)
    time_series_data = time_series_data.asfreq('D')  # Thiết lập tần số hàng ngày (có thể thay đổi tùy theo dữ liệu thực tế)
    
    # Chia dữ liệu thành tập huấn luyện và tập kiểm tra
    train_size = int(len(time_series_data) * 0.8)
    train, test = time_series_data.iloc[:train_size], time_series_data.iloc[train_size:]
     # Tìm các tham số tốt nhất cho mô hình ARIMA
    best_params = find_best_arima_params(train)
    # Huấn luyện mô hình ARIMA
    model_arima = ARIMA(train, order=best_params)
    model_arima_fit = model_arima.fit()
    
    # Dự đoán trên tập kiểm tra
    start = len(train)
    end = len(train) + len(test) - 1
    predictions = model_arima_fit.predict(start=start, end=end, typ='levels')
    
    # Tính toán lỗi dự đoán
    mse = mean_squared_error(test, predictions)
    print(f'Mean Squared Error for {column}: {mse}')
    
    # Lưu kết quả vào dictionary
    results[column] = {
        'test': test,
        'predictions': predictions,
        'mse': mse
    }

    # Vẽ biểu đồ cho từng trường
    plt.figure(figsize=(14, 8))
    plt.plot(results[column]['test'].index, results[column]['test'], label=f'Test {column}')
    plt.plot(results[column]['predictions'].index, results[column]['predictions'], label=f'Predictions {column}', linestyle='--')
    plt.title(f'ARIMA Model Predictions for {column}')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from arch import arch_model
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Chọn cột 'truong_1' để phân tích chuỗi thời gian
time_series = data['truong_4']

# Chia dữ liệu thành tập huấn luyện (train) và tập kiểm tra (test)
train_size = int(len(time_series) * 0.8)
train, test = time_series[:train_size], time_series[train_size:]

# Khởi tạo và fit mô hình GARCH
model = arch_model(train, vol='GARCH', p=1, q=1)
garch_fit = model.fit()

# Dự báo cho tập kiểm tra
garch_forecast = garch_fit.forecast(horizon=len(test))

# Lấy ra phần dự báo của chuỗi
garch_forecast_values = np.sqrt(garch_forecast.variance.values[-1, :])

# Tính các độ đo đánh giá
mae = mean_absolute_error(test, garch_forecast_values)
mse = mean_squared_error(test, garch_forecast_values)
rmse = np.sqrt(mse)

# In các độ đo đánh giá
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Vẽ biểu đồ chuỗi thời gian gốc, tập huấn luyện, tập kiểm tra và dự báo ARIMA
plt.figure(figsize=(12, 6))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(test.index, garch_forecast_values, label='GARCH Forecast', color='red')
plt.title('GARCH Model Forecast')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# In kết quả dự báo
print(garch_forecast_values)
```
